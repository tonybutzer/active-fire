```python
def normalize(array):
    """Normalizes numpy arrays into scale 0.0 - 1.0"""
    array_min, array_max = array.min(), array.max()
    return ((array - array_min)/(array_max - array_min))
```

```python
def plot_img(src):
    print("hello", src.shape)
    my_array = src.read(1)
    pyplot.figure()
    pyplot.imshow(my_array, cmap='gray')
    pyplot.show()
```

```python
def plot_hist(src):
    print("hello", src.shape)
    my_array = src.read(1)
    show_hist(my_array, bins=50, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled', title="Histogram")
```

```python
for file in file_list:
    full_name = '../testData/' + file
    print(full_name)
    with rasterio.open(full_name) as src:
        print(src.shape)
        print(src.crs)
        plot_hist(src)
```
