```python
def normalize(array):
    """Normalizes numpy arrays into scale 0.0 - 1.0"""
    array_min, array_max = array.min(), array.max()
    return ((array - array_min)/(array_max - array_min))
```

```python
def plot_img(src):
    print("hello", src.shape)
    my_array = src.read(1)
    pyplot.figure()
    pyplot.imshow(my_array, cmap='gray')
    pyplot.show()
```

```python
def plot_hist(src):
    print("hello", src.shape)
    my_array = src.read(1)
    show_hist(my_array, bins=50, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled', title="Histogram")
```

```python
import xarray as xr
# Copied from scotty h - dumb down for tony
def create_multiband_dataset(bands):
    '''A function to load multiple landsat bands into an xarray dataset '''
    
    # Each image is a dataset containing both band4 and band5
    datasets = []
    for band in bands:
        url = full_name = '../testData/' + band + '.tif'
        da = xr.open_rasterio(url)
        da = da.squeeze().drop(labels='band')
        ds = da.to_dataset(name=band)
        datasets.append(ds)

    DS = xr.merge(datasets)
    
    return DS
```

```python
def snip_tif(n): 
    return n.replace('.tif','')
  
bands = list(map(snip_tif, file_list)) 

bands
```

```python
big_xarray = create_multiband_dataset(bands)
```

```python
big_xarray
```

```python
# compute simple ndvi
```

```python
ds = big_xarray
NDVI = (ds['band5'] - ds['band4']) / (ds['band5'] + ds['band4'])
NDVI
```
